# **Concatenating DataFrames**

Concatenating DataFrames in pandas means stacking DataFrames together â€” either vertically (rows) or horizontally (columns).

pd.concat() is used for this.

```python
import pandas as pd
```

```python
# pd.concatenate([df1, df2], axis=1, ignore_index=True)  # axis=1 for columns, axis=0 for rows; ignore_index to reset index in new df
df1 = pd.DataFrame({
    "Name": ["Hayyan", "Haroon"],
    "Age": [19, 29]
})

df2 = pd.DataFrame({
    "Name": ["Rehman", "Fatima"],
    "Age": [27, 34]
})

result = pd.concat([df1, df2], axis=1, ignore_index=True)
print(result)
```
