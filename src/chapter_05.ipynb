```python
import pandas as pd
```

**Sorting and Aggregation**

sorting is used to rearrange rows in a DataFrame based on the values of one or more columns.

```python
data = {
    "Name":["Hayyan", "Haroon", "Khadija"],
    "Age":[19, 29, 25 ],
    "Salary":[80000, 50000, 5400 ]
}

df= pd.DataFrame(data)
df.sort_values(by=["Salary", "Age"], ascending=[True, True], inplace=True)
print(df)
```

**Aggregation**
summarizing data â†’ producing summary statistics.

| Function   | Description            |
| ---------- | ---------------------- |
| `sum()`    | Adds up all values     |
| `mean()`   | Computes average       |
| `median()` | Finds median           |
| `min()`    | Minimum value          |
| `max()`    | Maximum value          |
| `count()`  | Counts non-null values |
| `std()`    | Standard deviation     |
| `var()`    | Variance               |


```python
data = {
    "Name":["Hayyan", "Haroon", "Khadija", "Fatima", "Rehman"],
    "Age":[19, 19, 25, 34, 27 ],
    "Salary":[80000, 50000, 54000, 72000, 58000 ]
}
df= pd.DataFrame(data)

grouped = df.groupby("Age")["Salary"].sum()
print(grouped)
```
