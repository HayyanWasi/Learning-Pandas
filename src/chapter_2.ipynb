### **Filtering Rows based on certain conditions**.
used for example we want emplyee with salary under or above 50k, we will filter rows on basis of conditions

we are going to learn how we can solve bellow problems

- select specific column
- filter rows
- combine multiple conditions

if we want to select any col we can use square [] brackets.

for rows we will use Boolean Conditions

**Selecting columns**:
1. A series (single col of data)
2. Dataframe (Multiple cols of data)

**Filtering rows**:
1. extracting rows which met our any soecific condition 
*for example*: we want emplyee with salary under or above 50k

2. use boolean indexing

```python
#For cols

# print(df)
print("Single Column")
print(df['Name'])
```

```python
#selecting multiple columns
subset = df[["Name", "Salary"]]
print('Subset with multiple columns:')
print(subset)
```

**Filtering Row**

```python
high_salary =df[df["Salary"] > 50000]
print("Employees with salary greater than 50000")
print(high_salary)
```

```python
#Filtering based on Age  and Salary at same time
filtered_df = df[(df["Age"] > 30) & (df["Salary"] > 55000)]
print("Employees older than 30 with salary greater than 55000")
print(filtered_df)
```

```python
#Using OR condition -> return if only one condition is true
filtered_or = df[(df['Age']>35) | (df['Performance Score']>90)]
print("Employees older than 35 or with performance score greater than 90")
print(filtered_or)
```
