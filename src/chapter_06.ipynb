# **Merging and Joining DataFrame**

merging = merge 2 or more rows of diff dataframe based on common key column


| Join Type | What It Does                           |
| --------- | -------------------------------------- |
| **inner** | Matches only common keys               |
| **left**  | All left rows + matching right rows    |
| **right** | All right rows + matching left rows    |
| **outer** | All rows from both sides               |
| **cross** | Every combination of left Ã— right rows |


```python
# pd.merge(df1, df2, on="common_column", how="type of join") 

df_customers = pd.DataFrame({
    'CustomerID':[1,2,3],
    'Name':["Hayyan", "Abdullah", "Rehman"]
})

df_orders = pd.DataFrame({
    'CustomerID':[1,2,4],
    'OrderAmount':[250, 450, 350]
})

df_merged = pd.merge(df_customers, df_orders, on="CustomerID", how='inner')
print('right join')
print(df_merged)
```
