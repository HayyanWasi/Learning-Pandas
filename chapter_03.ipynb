# Adding Rows

```python
import pandas as pd
```

```python
data={
    "Name":["Ali", "Mehmood", "Akram", "Daniyal", "Habib", "Zohran", "Ghafoor", "Rehman"],
    "Age":[29, 34, 22, 38, 29, 45, 53, 60],
    "Salary":[50000, 60000, 45000, 52000, 49000, 70000, 48000, 58000],
    "Performance Score":[85, 90, 78, 92, 88, 95, 80, 54]
}

df=pd.DataFrame(data)
```

```python
df["Bonus"] = df["Salary"]*0.10
print(df)
```

insert() is a method used to add a new column at a specific position inside a DataFrame.

```python
df.insert(0, "Employee ID", [1,2,3,4,5,6,7,8])
print(df)
```

### Updating Values in DataFrame

.loc in Pandas is used to select rows and columns by labels (names), not by position.

```python
df.loc[0, "Salary"] = 55000
print(df)
```

```python
#increasing salary in same col

df['Salary'] = df['Salary'] * 1.1
print(df)
```
